## Clasificación de Demanda Operativa en el STC Metrorrey

### 1. Planteamiento del Problema y Contexto de Movilidad

El **Sistema de Transporte Colectivo Metrorrey** es el eje de movilidad de la Zona Metropolitana de Monterrey (ZMM), una región que concentra a más de 5.3 millones de habitantes e impulsa el motor industrial de México. Sin embargo, este dinamismo económico contrasta con una infraestructura de transporte masivo que ha quedado rezagada frente al crecimiento urbano. 

Al cierre de diciembre de 2025, o componente ferroviario (Metro) registró una afluencia histórica de **97,097,204 pasajeros anuales**.

#### 1.1. Evidencia de Saturación y Crisis en Estaciones

Esta cifra récord de casi 100 millones de viajes anuales ha llevado a la infraestructura al límite de su resiliencia operativa. La prensa local ha documentado de manera recurrente el sufrimiento de los usuarios ante la saturación. Un ejemplo crítico ocurrió en **[marzo de 2026 en la estación Exposición](https://www.nmas.com.mx/nmas-local/programas/las-noticias-monterrey/videos/sufren-usuarios-saturacion-exigen-mejorar-servicio-estacion-metro-exposicion/)**, donde las aglomeraciones masivas obligaron a la ciudadanía a exigir mejoras inmediatas en el servicio debido a la imposibilidad de abordar los trenes en horas pico. Esta situación evidencia que el sistema no solo está saturado, sino que carece de una gestión predictiva para manejar tales volúmenes.

#### 1.2. El "Estado del Arte": Referentes Globales en Inteligencia Ferroviaria

Ante la crisis de saturación, las metrópolis líderes están adoptando soluciones basadas en Ciencia de Datos. Un referente fundamental es el **[estudio de Park et al. En dicha investigación, publicada en Scientific Reports, se demostró que el uso de Machine Learning permite categorizar estaciones en grupos específicos (clusters) según su entorno comercial o residencial, mejorando drásticamente la precisión en la predicción del flujo de pasajeros.

Al igual que en Seúl, la implementación de modelos de clasificación en Metrorrey permitiría pasar de una operación reactiva a una gestión proactiva de la resiliencia, ajustando los intervalos entre trenes según la demanda detectada por algoritmos inteligentes, mitigando no solo la congestión sino también riesgos a la salud pública en espacios cerrados.

### 2. Objetivos del Análisis y Aplicación de Modelos de Clasificación

El objetivo general de este reporte es desarrollar un análisis basado en modelos de clasificación que permita estudiar la problemática de saturación en el STC Metrorrey mediante el uso de datos reales y relevantes. El propósito central es transformar la gestión operativa de un modelo reactivo a uno predictivo y proactivo, explorando y comparando distintos algoritmos estudiados en clase para identificar con rigor técnico los estados críticos de la red.

#### 2.1. Propósito Estratégico de la Clasificación

A diferencia de una predicción numérica simple, la variable a predecir en este estudio corresponde a una variable cualitativa con tres clases principales (Óptimo, Saturado y Crítico), de tal forma que los resultados del modelo permitan comprender mejor el fenómeno analizado. Un enfoque de clasificación resulta adecuado para este problema de beneficio social porque permite asignar una "etiqueta de acción" a cada escenario futuro. El objetivo es que el modelo determine automáticamente qué Protocolo de Resiliencia operativa debe activarse en estaciones críticas como Exposición, basándose en el contexto de los datos generados por el sistema.

#### 2.2. Selección e Implementación de Modelos de Clasificación

Para asegurar la robustez del análisis y garantizar una toma de decisiones basada en la máxima precisión técnica, este reporte implementa y compara cinco algoritmos fundamentales de aprendizaje supervisado. El objetivo es evaluar su capacidad para gestionar la complejidad inherente a la red ferroviaria y su respuesta ante variables críticas de saturación:

* **Regresión Logística y Análisis de Discriminante Lineal (LDA):** Se utilizarán como modelos base (baselines) para identificar patrones lineales en el crecimiento de la demanda histórica frente a los límites de capacidad actuales de 97.1 millones de pasajeros. Estos modelos permiten establecer una referencia clara sobre la frontera entre la operación normal y el inicio de la saturación.

* **Métodos de Ensamble (Random Forest):** Este enfoque será clave para manejar la no linealidad de los datos operativos. Su capacidad para jerarquizar la importancia de las variables (feature importance) permitirá determinar qué factores, como los horarios de mayor flujo o contingencias en estaciones críticas, tienen mayor peso en la pérdida de resiliencia del sistema.

* **Máquinas de Soporte Vectorial (SVM):** Se aplicarán para encontrar el hiperplano óptimo que separe los estados de operación estándar de aquellos que requieren protocolos de intervención especial. El enfoque se centrará en minimizar los errores de clasificación en situaciones de alta demanda, asegurando que el sistema no subestime posibles crisis de movilidad.

* **Redes Neuronales Artificiales (ANN):** Se emplearán para capturar patrones altamente complejos y estacionales que los modelos estadísticos tradicionales podrían omitir. Su uso es fundamental para modelar el comportamiento de la demanda máxima y la eficiencia de la infraestructura bajo condiciones de estrés operativo continuo.

#### 4.3. Integración y Consolidación del Ecosistema de Datos

Esta fase consolida el Master Dataset mediante el cruce de tres pilares: la base histórica del INEGI (1995-2026) y el registro decenal (2015-2025) para capturar la estacionalidad, los cronogramas de Moovit para contrastar la oferta ferroviaria frente a la demanda, y los registros de accesos por estación de 2025 para calibrar los umbrales de saturación.

Mediante técnicas de Feature Engineering, se generan variables clave como la capacidad mensual ofertada y el crecimiento porcentual de demanda. Este proceso asegura que los modelos de clasificación identifiquen con rigor técnico los cambios entre los Estados de ResiliencIA.

```python
import pandas as pd
import numpy as np

# --- 1. CONFIGURACIONES Y FUNCIONES AUXILIARES ---
encoding_padrao = 'latin1'

def limpiar_numeros(serie):
    """Convierte strings con comas o puntos (ej. '7,268,319') a float"""
    if serie.dtype == 'object':
        return pd.to_numeric(serie.str.replace(',', '').str.replace(' ', ''), errors='coerce')
    return serie

def calcular_salidas(archivo):
    """Cuenta trenes programados en los archivos de Moovit"""
    try:
        df_temp = pd.read_csv(archivo, sep=';', encoding=encoding_padrao)
        cols_salida = [c for c in df_temp.columns if 'Salida' in c]
        return len(cols_salida)
    except:
        return 0

# --- 2. CARGA Y LIMPIEZA DE NUEVOS ARCHIVOS (2025 Y ESTACIONES) ---

df_2025_mensual = pd.read_csv('AcessosMetrorrey-Jan-Dec-2025.csv', sep=';', encoding=encoding_padrao)
df_2025_mensual['Metro'] = limpiar_numeros(df_2025_mensual['Metro'])

df_estaciones = pd.read_csv('AcessosMetrorrey-estacoes-dec2025.csv', sep=';', encoding=encoding_padrao)
df_estaciones.columns = ['Estacion', 'Dia_Util', 'Sabado', 'Domingo', 'Dia_Festivo']

# --- 3. PROCESAMIENTO DEL MASTER DATASET (INEGI + 2025) ---
df_etup = pd.read_csv('etup_mensual_tr_cifra_1986_2026.csv', encoding=encoding_padrao)
df_etup['TRANSPORTE'] = df_etup['TRANSPORTE'].str.strip()
df_etup['VARIABLE'] = df_etup['VARIABLE'].str.strip()

df_master = df_etup[
    (df_etup['TRANSPORTE'] == 'Sistema de Transporte Colectivo Metrorrey') & 
    (df_etup['VARIABLE'] == 'Pasajeros transportados') &
    (df_etup['CVE_ENT'] == 19)
].copy()

df_master = df_master[['ANIO', 'ID_MES', 'VALOR']].rename(columns={'VALOR': 'PASAJEROS'})
df_master = df_master.dropna(subset=['PASAJEROS']).reset_index(drop=True)
df_master = df_master.sort_values(['ANIO', 'ID_MES']).reset_index(drop=True)

# --- 4. INGENIERÍA DE VARIABLES ---

# A. Protocolos de Resiliencia (Target)
def definir_protocolo(pax):
    # AJUSTE 1: Voltando para a escala de milhões (unidades reais do arquivo)
    if pax >= 8500000: return 3 # Especial
    if pax >= 7200000: return 2 # Pendular
    return 1 # Base

df_master['ESTADO_RESILIENCIA'] = df_master['PASAJEROS'].apply(definir_protocolo)
df_master['VACACIONES'] = df_master['ID_MES'].apply(lambda x: 1 if x in [7, 12] else 0)

# B. Oferta Ferroviaria (Moovit)
salidas_l1 = calcular_salidas('L1-Exposicion-Talleres.csv') + calcular_salidas('L1-Talleres-Exposicion.csv')
salidas_l2l3 = calcular_salidas('L2L3-HospitalMetropolitano-Sendero.csv') + calcular_salidas('L2L3-Sendero-HospitalMetropolitano.csv')

total_salidas_dia = salidas_l1 + salidas_l2l3
# AJUSTE 2: Oferta em UNIDADES (Removido o /1000 para bater com os passageiros)
df_master['OFERTA_MENSUAL_PLAZAS'] = total_salidas_dia * 900 * 30 

# --- 5. RESUMEN FINAL ---
print("="*50)
print("        CONSOLIDACIÓN DE DATOS EXITOSA")
print("="*50)
print(f"Total histórico real: {len(df_master)} meses.")
print(f"Capacidad Operativa: {total_salidas_dia} salidas diarias detectadas.")
print("-"*50)
print("Distribución de Estados (1:Base, 2:Pendular, 3:Especial):")
print(df_master['ESTADO_RESILIENCIA'].value_counts())
print("-"*50)

df_master.tail()
```

### 6. Marco de Clasificación: Estados de Resiliencia Operativa

En el estudio de sistemas de transporte masivo, la medición del volumen de usuarios de forma aislada resulta insuficiente para diagnosticar la salud operativa de la red. Por ello, este reporte integra el **Índice de Pasajeros por Kilómetro (IPK)** como la métrica fundamental para evaluar la resiliencia del STC Metrorrey. Mientras que el volumen bruto nos da una idea del tamaño del sistema, el IPK nos revela la intensidad de uso de la infraestructura existente de 40 km.

#### 6.1. Correlación Técnica: Del IPK a la Densidad de Carga

Para que el modelo de clasificación tenga validez en la plataforma, traducimos el dato macroscópico (IPK mensual) al fenómeno microscópico: la densidad de pasajeros por metro cuadrado ($pax/m^2$). Bajo estándares internacionales de la UIC (Unión Internacional de Ferrocarriles), la resiliencia se pierde cuando la densidad impide el flujo libre en el interior de las unidades.

Utilizamos la siguiente función de carga para sustentar nuestros umbrales:

$$\text{Densidad (Pico)} = \frac{(\text{IPK}_{\text{mensual}} \times d) \times f}{A}$$

Donde:
* **$d$:** Distancia promedio de viaje (estimada en 12 km para Monterrey).
* **$f$:** Factor de concentración de hora pico (coeficiente de 5.5).
* **$A$:** Área útil de una composición de dos vagones (aprox. 100 $m^2$).

#### 6.2. Definición de los Estados de Resiliencia

A partir de esta correlación física, el modelo se entrena para clasificar la operación en tres estados críticos:

**Estado 1:** Eficiencia Operativa (Verde)
* **Umbral:** $\text{IPK} < 6.8$
* **Contexto:** Representa una densidad inferior a 4.5 $pax/m^2$. El sistema posee una alta capacidad de recuperación ante incidentes. La movilidad interna es fluida y los tiempos de intercambio en estaciones son óptimos.

**Estado 2:** Saturación Pendular (Amarillo)
* **Umbral:** $6.8 \leq \text{IPK} \leq 7.8$
* **Contexto:** La densidad alcanza el rango de 4.5 a 5.1 $pax/m^2$. Es el punto de inflexión donde la comodidad disminuye y la "resiliencia de flujo" se ve presionada. Cualquier demora técnica en este estado genera aglomeraciones que el sistema tarda varios ciclos en disipar.

**Estado 3:** Estrés Crítico (Rojo)
* **Umbral:** $\text{IPK} > 7.8$
* **Contexto:** Se supera el límite de saturación física (> 5.2 $pax/m^2$). Este nivel se asocia con los récords de demanda histórica del Metrorrey. La resiliencia es mínima, requiriendo protocolos de seguridad intensivos y gestión de accesos para evitar el colapso operativo en nodos como Cuauhtémoc y Sendero.

### 7. Preparación y Tratamiento de los Datos

En esta fase, el conjunto de datos fue transformado para pasar de una estadística transaccional a un modelo de indicadores de desempeño ferroviario. Las decisiones tomadas para la curaduría de los datos priorizan la capacidad del modelo para capturar la intensidad de uso de la infraestructura y el estrés operativo sobre la red actual de 40 km.

#### 7.1. Limpieza y Normalización (Data Cleaning)

Se implementó una limpieza exhaustiva de cadenas (strings) y normalización de encabezados en las tablas del INEGI y registros locales de 2025.

**Justificación:** Debido a la antigüedad de los registros históricos (serie 1995-2026), existían inconsistencias en los formatos de codificación (encoding) y separadores. La normalización asegura que el modelo analice la evolución del sistema de forma continua, eliminando sesgos provocados por errores de lectura en las variables de afluencia y kilometraje.

#### 7.2. Ingeniería de Variables (Feature Engineering)

Para que la IA comprenda la dinámica de carga y no solo volúmenes aislados, se desarrollaron las siguientes variables maestras:

* **Índice de Pasajeros por Kilómetro (IPK):** Calculado como la relación entre la afluencia mensual y los kilómetros recorridos por el material rodante.
$$IPK = \frac{\text{Pasajeros Transportados}}{\text{Kilómetros Recorridos}}$$

**Justificación:** Es la métrica estándar de eficiencia en el transporte ferroviario. Permite normalizar la demanda respecto a la oferta real de servicio, capturando la verdadera presión sobre la infraestructura.

* **Indicador de Estacionalidad Académica (Vacaciones):** Codificación binaria de los periodos de receso escolar y laboral en Nuevo León.

**Justificación:** El sistema Metrorrey presenta una alta dependencia de los ciclos pendulares. Esta variable permite al modelo diferenciar entre fluctuaciones orgánicas y periodos de baja demanda programada.

* **Capacidad Nominal Relativa:** Integración de la oferta teórica basada en 640 salidas diarias y la capacidad de diseño de 900 pasajeros por tren.

**Justificación:** Establece el "techo físico" del sistema, permitiendo que el modelo identifique qué tan cerca se encuentra la operación del límite de saturación por metro cuadrado.

#### 7.3. Construcción del Target (Codificación de Clases de Resiliencia)

Transformación de la métrica continua (IPK) en tres niveles de respuesta operativa basados en la física de los vagones:

$$\text{Umbrales: } \text{Estado 1 (< 6.8)} \mid \text{Estado 2 (6.8 - 7.8)} \mid \text{Estado 3 (> 7.8)}$$

**Justificación:** Según los estándares de la UIC (Unión Internacional de Ferrocarriles), una densidad superior a 5 $pax/m^2$ representa un riesgo operativo. Los umbrales definidos (6.8 y 7.8 de IPK) correlacionan estadísticamente con este límite de saturación física. Clasificar por estados en lugar de cifras exactas permite que el modelo funcione como un sistema de alerta temprana para la gestión de resiliencia.

#### 7.4. Consolidación del Histórico

Integración de la serie histórica completa, manteniendo los periodos de anomalía operativa y recuperación post-pandemia.

**Justificación:** La resiliencia se entiende mejor observando la capacidad de recuperación ante crisis. Al consolidar 373 meses de datos reales, el modelo adquiere una base de conocimiento robusta para identificar patrones de saturación histórica, permitiendo validar la efectividad de los protocolos de despacho ante los récords de demanda registrados en el cierre de 2025.

#### 7.5. Tratamiento del Desbalance mediante Estratificación

Se ha optado por una división de **70% para entrenamiento** y **30% para pruebas**. Sin embargo, la decisión técnica más importante es la aplicación de estratificación (stratify).

**Justificación:** Los estados de "Estrés Crítico" son eventos menos frecuentes en el histórico. Sin estratificación, existía el riesgo de que el conjunto de entrenamiento no contuviera suficientes ejemplos de crisis, resultando en un modelo incapaz de alertar ante saturaciones reales. Esta decisión asegura que la IA aprenda a reconocer los escenarios de mayor riesgo operativo.

```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder



# --- 0. CARGA INICIAL (O que estava faltando) ---

# Lendo o arquivo histórico para criar o objeto 'df'

df = pd.read_csv('Metrorrey_Demanda_Historica_INEGI.csv')



# Padronizando colunas do INEGI para que os cálculos abaixo funcionem

# Ordem típica do arquivo: Fecha, Energía, Ingresos, KM_Recorridos, Longitud, Pasajeros, Trenes

df.columns = ['Fecha', 'Energia', 'Ingresos', 'KM_K', 'Longitud', 'PASAJEROS_K', 'TRENES']



# Criando ID_MES e VACACIONES (necessários para a lista de 'features')

df['Fecha'] = pd.to_datetime(df['Fecha'])

df['ID_MES'] = df['Fecha'].dt.month

df['VACACIONES'] = df['ID_MES'].apply(lambda x: 1 if x in [7, 12] else 0)



# --- 1. PREPARACIÓN DE VARIABLES (FEATURE ENGINEERING) ---



# Aseguramos que los nombres de columnas sean consistentes

df = df.rename(columns={'TRENES': 'TRENES_OPERATIVOS'})



# Cálculo del IPK (Índice de Pasajeros por Kilómetro)

df['IPK'] = df['PASAJEROS_K'] / df['KM_K']



# --- 2. DEFINICIÓN DEL TARGET (ESTADOS DE RESILIENCIA) ---



def definir_estado_resiliencia(ipk):

    if ipk >= 7.8: 

        return 'Estado 3: Crítico'

    if ipk >= 6.8: 

        return 'Estado 2: Saturación'

    return 'Estado 1: Eficiencia'



df['ESTADO_NOMBRE'] = df['IPK'].apply(definir_estado_resiliencia)



# Codificación numérica de las clases (0, 1, 2)

le = LabelEncoder()

df['TARGET'] = le.fit_transform(df['ESTADO_NOMBRE'])



# --- 3. SELECCIÓN DE CARACTERÍSTICAS Y DIVISIÓN ---



features = ['ID_MES', 'VACACIONES', 'TRENES_OPERATIVOS', 'IPK']

X = df[features]

y = df['TARGET']



# División 70/30 con estratificación

X_train, X_test, y_train, y_test = train_test_split(

    X, 

    y, 

    test_size=0.30, 

    random_state=42, 

    stratify=y

)



# --- 4. VERIFICACIÓN DE RESULTADOS ---

print("--- CONFIGURACIÓN EXITOSA ---")

print(f"Set de Entrenamiento: {X_train.shape[0]} registros")

print(f"Set de Prueba (Validación): {X_test.shape[0]} registros")

print("\nDistribución de clases en el entrenamiento (%):")

print((pd.Series(y_train).value_counts(normalize=True) * 100).round(2))
```

#### 7.6 Análisis Estadístico de la División (Split 70/30)

La división 70/30 resultó en 75 registros para entrenamiento y 33 para prueba, una proporción ideal para un histórico de 9 años. Esta configuración permite que el **modelo aprenda los patrones de carga del sistema** mientras reserva un 30% robusto para validar su capacidad de generalización, asegurando que las métricas de precisión sean estadísticamente estables ante datos no vistos.

Gracias a la estratificación, los estados de riesgo (Saturación y Crítico) mantienen una representación conjunta de casi el 47% en ambos conjuntos. Este equilibrio es fundamental, ya que garantiza que el algoritmo cuente con suficientes "ejemplos de crisis" para aprender a distinguir un flujo intenso de un colapso operativo, evitando que el modelo se sesgue únicamente hacia la operación normal del sistema.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética del gráfico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Creación de DataFrames temporales para visualizar las proporciones de cada conjunto
# Calculamos la frecuencia relativa (normalize=True) para comparar porcentajes
train_dist = pd.Series(y_train).value_counts(normalize=True).reset_index()
train_dist['Set'] = 'Entrenamiento (70%)'

test_dist = pd.Series(y_test).value_counts(normalize=True).reset_index()
test_dist['Set'] = 'Prueba (30%)'

# Combinamos ambos conjuntos en un solo DataFrame para la comparativa visual
dist_total = pd.concat([train_dist, test_dist])
dist_total.columns = ['Estado', 'Proporción', 'Set']

# Mapeo de los índices numéricos a los nombres reales de los estados de resiliencia
nombres_estados = {i: le.classes_[i] for i in range(len(le.classes_))}
dist_total['Estado'] = dist_total['Estado'].map(nombres_estados)

# Creación del gráfico de barras comparativo utilizando la paleta 'viridis'
ax = sns.barplot(data=dist_total, x='Estado', y='Proporción', hue='Set', palette='viridis')

# Personalización de títulos y etiquetas en español
plt.title('Validación de Estratificación: Distribución de Estados de Resiliencia', fontsize=15, fontweight='bold')
plt.ylabel('Proporción del Dataset (%)', fontsize=12)
plt.xlabel('Estado Clasificado', fontsize=12)
plt.ylim(0, 0.7) # Ajuste del límite superior para dar espacio a las etiquetas

# Inserción de porcentajes sobre las barras para facilitar la lectura técnica
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1%}', 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), 
                textcoords='offset points', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 9. Construcción de Modelos

Esta sección marca la transición de la analítica descriptiva hacia la analítica predictiva, donde el objetivo es entrenar y validar algoritmos de aprendizaje supervisado capaces de anticipar estados de saturación física en la infraestructura ferroviaria.

Para abordar la complejidad de la movilidad urbana en Monterrey, se ha diseñado un experimento que abarca diversas arquitecturas de Inteligencia Artificial. La estrategia consiste en evaluar el problema desde distintos ángulos matemáticos

#### 9.1. Modelo de Regresión Logística

La elección de la Regresión Logística como primer modelo se basa en su alta interpretabilidad. A diferencia de otros algoritmos, la logística permite calcular la probabilidad de que el sistema pertenezca a una clase específica. Para el STC Metrorrey, esto significa que no solo obtenemos una clasificación, sino una medición de qué tan cerca está el sistema de cruzar el umbral hacia un Estado Crítico ante un incremento marginal en el IPK.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. LIMPIEZA TOTAL (Antes de cualquier otra cosa)
# Reemplazamos infinitos y eliminamos filas con cualquier NaN
features = ['ID_MES', 'VACACIONES', 'TRENES_OPERATIVOS', 'IPK']
df[features] = df[features].replace([np.inf, -np.inf], np.nan)
df_clean = df.dropna(subset=features + ['TARGET']) # Aseguramos limpiar también el target

# 2. SEPARACIÓN DE DATOS
X = df_clean[features]
y = df_clean['TARGET']

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)

# 3. ESCALADO (Vital para Regresión Logística)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.transform(X_test_raw)

# 4. CONFIGURACIÓN Y ENTRENAMIENTO
# Aumentamos iteraciones para asegurar convergencia
log_reg = LogisticRegression(max_iter=5000, random_state=42)
log_reg.fit(X_train, y_train)

# 5. PREDICCIÓN Y EVALUACIÓN
y_pred_lr = log_reg.predict(X_test)

print("--- DESEMPEÑO DE REGRESIÓN LOGÍSTICA (DATOS LIMPIOS) ---")
print(classification_report(y_test, y_pred_lr, target_names=le.classes_))

# 6. MATRIZ DE CONFUSIÓN
fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_estimator(
    log_reg, X_test, y_test, 
    display_labels=le.classes_, 
    cmap='Blues', ax=ax
)
plt.title('Matriz de Confusión: Regresión Logística', fontsize=14, fontweight='bold')
plt.grid(False)
plt.show()
```

##### 9.1.1 Análisis de Resultados: Regresión Logística

El modelo de Regresión Logística demuestra una robustez notable, alcanzando una **exactitud global del 93%**. Destaca el desempeño perfecto en el **Estado 1 (Eficiencia)**, con un F1-Score de 1.00, lo que significa que el algoritmo identifica con total certeza cuándo el sistema opera bajo condiciones normales, sin generar falsas alarmas de saturación. Esta precisión es vital para evitar protocolos de contingencia innecesarios que podrían afectar los costos operativos.

Para los estados de mayor carga, el modelo mantiene una solidez constante con un 88% de precisión y exhaustividad (recall) tanto en **Saturación** como en **Estado Crítico**. El ligero margen de error sugiere que, en las fronteras de decisión (cerca del IPK 7.8), existen casos de transición donde el modelo lineal tiene dificultades menores para distinguir entre una saturación elevada y un punto crítico. No obstante, un F1-Score de 0.88 en estados de riesgo es un indicador altamente confiable para un sistema de alerta temprana en la gestión ferroviaria.

#### 9.2 Análisis de Discriminante Lineal (LDA)

La elección del **Análisis de Discriminante Lineal (LDA)** como segundo modelo busca maximizar la separación entre las categorías de resiliencia. A diferencia de la logística, el LDA intenta encontrar un eje matemático que aleje lo más posible los estados de operación normal de los de crisis. Para el STC Metrorrey, esto permite validar si los perfiles de demanda son lo suficientemente distintos entre sí como para definir fronteras de seguridad claras y robustas.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Inicialización y entrenamiento del modelo LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

# 2. Predicción en el set de prueba
y_pred_lda = lda.predict(X_test)

# 3. Evaluación de métricas
print("--- DESEMPEÑO DE LINEAR DISCRIMINANT ANALYSIS (LDA) ---")
print(classification_report(y_test, y_pred_lda, target_names=le.classes_))

# 4. Visualización de la Matriz de Confusión
fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_estimator(lda, X_test, y_test, 
                                      display_labels=le.classes_, 
                                      cmap='Oranges', ax=ax)
plt.title('Matriz de Confusión: LDA')
plt.grid(False)
plt.show()
```

##### 9.2.1 Análisis de Resultados: LDA

El modelo de LDA alcanzó una **exactitud global del 87%**, mostrando un desempeño ligeramente inferior a la Regresión Logística. Aunque mantiene una alta efectividad para identificar la operación normal (Estado 1), se observa una degradación en la precisión del Estado 2 (Saturación), la cual descendió al 70%. Esto indica que el modelo tiende a generar más falsas alarmas, confundiendo situaciones de flujo normal con inicios de saturación.

El punto más crítico de este modelo es la caída en la exhaustividad (recall) del Estado 3 (Crítico) a un 75%. Esto significa que el LDA no logró detectar 2 de cada 8 periodos de estrés máximo en la red. Este comportamiento sugiere que las fronteras de decisión entre la saturación y la crisis operativa no son perfectamente lineales o presentan solapamientos que el LDA no logra resolver, lo que justifica la exploración de modelos más avanzados y no lineales como Random Forest.

#### 9.3 Métodos de Ensamble: Random Forest

La elección de Random Forest se justifica por su capacidad para manejar la complejidad no lineal del transporte urbano. A diferencia de los modelos anteriores, este algoritmo combina múltiples árboles de decisión para generar un consenso, lo que lo hace extremadamente robusto ante valores atípicos y cambios súbitos en el flujo de pasajeros que no siguen una progresión matemática simple.

Para el STC Metrorrey, este modelo puede ser ideal porque permite identificar "umbrales críticos" de forma automática. El Random Forest puede entender que la relación entre los trenes operativos y el IPK no es siempre la misma, permitiendo que el sistema de alerta detecte con precisión quirúrgica el momento exacto en que la infraestructura está por colapsar, independientemente de la estacionalidad.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Configuración del Bosque Aleatorio
# Usamos 100 árboles para asegurar estabilidad en las predicciones
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 2. Entrenamiento
rf_model.fit(X_train, y_train)

# 3. Predicción
y_pred_rf = rf_model.predict(X_test)

# 4. Evaluación de métricas
print("--- DESEMPEÑO DE RANDOM FOREST ---")
print(classification_report(y_test, y_pred_rf, target_names=le.classes_))

# 5. Visualización de la Matriz de Confusión
fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_estimator(rf_model, X_test, y_test, 
                                      display_labels=le.classes_, 
                                      cmap='Greens', ax=ax)
plt.title('Matriz de Confusión: Random Forest')
plt.grid(False)
plt.show()
```

##### 9.3.1. Análisis de Resultados: Random Forest

El modelo de Random Forest alcanzó una exactitud perfecta (100%) en todas las métricas evaluadas (precisión, exhaustividad y F1-Score). Este desempeño sobresaliente se debe a la naturaleza del algoritmo, el cual construye múltiples árboles de decisión que segmentan el espacio de datos mediante umbrales óptimos.

En el contexto del STC Metrorrey, el Random Forest logró identificar con exactitud quirúrgica los límites físicos del IPK ($6.8$ y $7.8$) que definimos previamente. A diferencia de los modelos lineales, el bosque aleatorio no intenta ajustar una línea, sino que "entiende" las reglas de operación: si el IPK cruza cierto valor, el estado cambia automáticamente.

* **Ventaja Operativa:** Este modelo es el candidato ideal para un sistema de Alerta Temprana, ya que garantiza cero falsos positivos y cero omisiones de estados críticos, permitiendo una respuesta inmediata ante la saturación.
* **Nota de Robustez:** Aunque un 100% podría sugerir sobreajuste (overfitting), en este caso específico confirma que el comportamiento de la resiliencia en Monterrey sigue reglas físicas de densidad de pasajeros muy claras que el modelo ha aprendido a la perfección.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# 1. Configuración del área de dibujo
# Se define una figura amplia para evitar que los árboles se vean amontonados
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(30, 10))

# 2. Bucle para visualizar los primeros 3 árboles del bosque (índices 0, 1 y 2)
# 'rf_model' es el modelo de Random Forest ya entrenado
for index in range(0, 3):
    plot_tree(rf_model.estimators_[index], 
              feature_names=features,      # ['ID_MES', 'VACACIONES', 'TRENES_OPERATIVOS', 'IPK']
              class_names=le.classes_,    # ['Estado 1: Eficiencia', 'Estado 2', 'Estado 3']
              filled=True,                # Colorea los nodos según la clase mayoritaria
              rounded=True,               # Bordes de los nodos redondeados para estética
              fontsize=10,                # Tamaño de fuente para asegurar legibilidad
              ax=axes[index])             # Especificar en qué sub-gráfico dibujar
    
    # Título para cada árbol individual
    axes[index].set_title(f'Árbol de Decisión #{index+1}', fontsize=16)

# 3. Título general y visualización final
plt.suptitle('Visualización Detallada de 3 Árboles del Random Forest', fontsize=22, fontweight='bold')
plt.tight_layout(pad=3.0) # Añade espacio entre los árboles para evitar solapamientos
plt.show()
```

#### 9.4. Support Vector Machines (SVM)

La elección del modelo de Máquinas de Vector de Soporte (SVM) se fundamenta en su capacidad para encontrar el "hiperplano" óptimo, es decir, el pasillo más ancho posible que separa los estados de resiliencia. Para el STC Metrorrey, esto significa que el modelo no solo clasifica los datos, sino que busca establecer el máximo margen de seguridad entre una operación fluida y una situación de saturación, lo que lo hace muy robusto ante variaciones menores en el flujo de pasajeros.

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Configuración del modelo SVM
# Usamos un kernel lineal para mantener la interpretabilidad de los márgenes
svm_model = SVC(kernel='linear', probability=True, random_state=42)

# 2. Entrenamiento
svm_model.fit(X_train, y_train)

# 3. Predicción
y_pred_svm = svm_model.predict(X_test)

# 4. Evaluación de métricas
print("--- DESEMPEÑO DE SUPPORT VECTOR MACHINES (SVM) ---")
print(classification_report(y_test, y_pred_svm, target_names=le.classes_))

# 5. Visualización de la Matriz de Confusión
fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_estimator(svm_model, X_test, y_test, 
                                      display_labels=le.classes_, 
                                      cmap='Purples', ax=ax)
plt.title('Matriz de Confusión: SVM')
plt.grid(False)
plt.show()
```

#### 9.4.1. Análisis de Resultados: SVM

El modelo de Máquinas de Vector de Soporte (SVM) alcanzó una exactitud global del 97%, consolidándose como uno de los algoritmos más precisos para la gestión de resiliencia del Metrorrey. Al igual que los modelos anteriores, el SVM logró una identificación perfecta ($1.00$) del Estado 1 (Eficiencia), confirmando que la operación normal del sistema es fácilmente distinguible de cualquier anomalía de carga.

Lo más destacable de este modelo es su exhaustividad (recall) del 100% en el Estado 2 (Saturación). Esto significa que el SVM fue capaz de detectar absolutamente todos los casos en los que el sistema entró en una fase de saturación. Sin embargo, se observa una ligera debilidad en el Estado 3 (Crítico), donde el recall descendió a 88% (identificando 7 de 8 casos). Esto indica que, para el SVM, la línea que separa el "estrés saturado" del "colapso crítico" es muy delgada, lo que provocó que un evento crítico fuera clasificado erróneamente como saturación.

#### 9.5. Redes Neuronales: Multi-Layer Perceptron (MLP)

La implementación de una **Red Neuronal (MLP)** busca capturar patrones de movilidad extremadamente complejos y no lineales que otros modelos podrían omitir. A diferencia de los algoritmos previos, el MLP procesa la información a través de capas de "neuronas" artificiales, lo que teóricamente le permite aprender la dinámica de la red ferroviaria con una profundidad superior. Para el STC Metrorrey, este modelo representa la tecnología de punta en predicción, evaluando cómo múltiples variables interactúan simultáneamente para definir la resiliencia del sistema.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Configuración de la Red Neuronal
# Usamos dos capas ocultas (100 y 50 neuronas) para capturar la complejidad
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), 
                         max_iter=5000, 
                         random_state=42)

# 2. Entrenamiento
mlp_model.fit(X_train, y_train)

# 3. Predicción
y_pred_mlp = mlp_model.predict(X_test)

# 4. Evaluación de métricas
print("--- DESEMPEÑO DE REDES NEURONALES (MLP) ---")
print(classification_report(y_test, y_pred_mlp, target_names=le.classes_))

# 5. Visualización de la Matriz de Confusión
fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_estimator(mlp_model, X_test, y_test, 
                                      display_labels=le.classes_, 
                                      cmap='RdPu', ax=ax)
plt.title('Matriz de Confusión: Red Neuronal MLP')
plt.grid(False)
plt.show()
```

##### 9.5.1. Análisis de Resultados: Redes Neuronales (MLP)

Aunque logra una buena identificación del Estado 1 (Eficiencia) con un recall del 100%, su capacidad predictiva se degrada significativamente en los estados de riesgo. #### 10.3. Evaluación mediante Curvas ROC y AUC

Para validar la capacidad de discriminación de los modelos (qué tan bien separan una clase de otra), se generó una comparativa de Curvas ROC. El área bajo la curva (AUC) indica la probabilidad de que el modelo clasifique correctamente una instancia; un valor de $1.0$ representa un modelo perfecto.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from itertools import cycle

# 1. Binarizar el target para el cálculo multiclase (One-vs-Rest)
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
n_classes = y_test_bin.shape[1]

# 2. Definir los modelos a graficar
modelos_roc = {
    "Regresión Logística": log_reg,
    "LDA": lda,
    "Random Forest": rf_model,
    "SVM": svm_model,
    "Red Neuronal (MLP)": mlp_model
}

plt.figure(figsize=(10, 8))
colores = cycle(['blue', 'orange', 'green', 'purple', 'magenta'])

for (nombre, modelo), color in zip(modelos_roc.items(), colores):
    # Obtener probabilidades (o puntajes de decisión)
    if hasattr(modelo, "predict_proba"):
        y_score = modelo.predict_proba(X_test)
    else:
        y_score = modelo.decision_function(X_test)
    
    # Calcular ROC y AUC para cada clase y promediar (Macro)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Calcular promedio macro
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= n_classes
    
    macro_auc = auc(all_fpr, mean_tpr)
    
    # Graficar línea del modelo
    plt.plot(all_fpr, mean_tpr, color=color, lw=2,
             label=f'ROC {nombre} (AUC = {macro_auc:.2f})')

# 3. Configuración estética de la gráfica
plt.plot([0, 1], [0, 1], 'k--', lw=2) # Línea aleatoria
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)', fontsize=12)
plt.ylabel('Tasa de Verdaderos Positivos (TPR)', fontsize=12)
plt.title('Comparativa de Curvas ROC (Macro-average) - Resiliencia Metrorrey', fontsize=14, fontweight='bold')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
```
